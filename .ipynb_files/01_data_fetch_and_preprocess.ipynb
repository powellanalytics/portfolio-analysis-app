```python
import yfinance as yf
import pandas as pd
from bs4 import BeautifulSoup
import requests
import datetime

import sys
sys.path.append('../utils')
import clean_column_names
```

```python
## Read S&P 500 tickers from Wikipedia and store in a list

request = requests.get(
'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = BeautifulSoup(request.content)
main_table = soup.find(id='constituents')
table = main_table.find('tbody').findAll('tr')
table = table[1:]

tickers = []

for row in table:
    ticker = row.findAll('td')[0].text
    tickers.append(ticker)

tickers = [s.replace('\n', '') for s in tickers]
```

```python
tickers
```

```python
## Add S&P 500 Index to the tickers list
tickers_with_sp500 = tickers + ['^GSPC']
tickers_with_sp500
```

```python
# ## Download price time series data from Yahoo Finance using yfinance (including S&P 500 Index)

start = datetime.datetime(2020, 1, 1)
# end = datetime.datetime(2022, 1, 1) -- not needed as we are downloading till today
# data = yf.download(tickers, start=start, end=end) -- not needed as we are downloading till today
price_data = yf.download(tickers_with_sp500, start=start)
```

```python
## View the last 5 rows of the data. You can see that the data is in a multi-index format.
price_data.tail()
```

```python
## We will only need the Adj Close column for our analysis. We can also convert the multi-index format to a single index format using the reset_index() function.
adj_close = price_data['Adj Close'].reset_index()
## View the last 5 rows of the data. You can see that the data is in a single index format.
adj_close.tail()
```

```python
## We can also set the Date column as the index using the set_index() function.
adj_close = adj_close.set_index('Date')
adj_close.tail()
```

```python
## Let's convert the daily prices to daily returns using the pct_change() function.
returns = adj_close.pct_change()
returns.tail()
```

```python
# Separate the '^GSPC' column
sp500_returns = returns[['^GSPC']].rename(columns={'^GSPC': 'RETURN'})
sp500_returns['TICKER'] = "SP500"
sp500_returns.tail()
```

```python
# Drop the '^GSPC' column from the original DataFrame for stacking
returns_data_excluding_sp500 = returns.drop(columns=['^GSPC'])
returns_data_excluding_sp500.tail()
```

```python
# Stack, reset index, and sort as before
stacked_returns = returns_data_excluding_sp500.stack(level=-1, dropna=False).reset_index().rename(index=str, columns={"level_1": "TICKER", 0:"RETURN"}).sort_values(['TICKER','Date']).set_index('Date')
stacked_returns.tail()
```

```python
# Merge the 'XYZ' column back
# Assuming 'Date' is the index of sp500_price, if not, you might need to reset and rename its index as well
combined_returns = stacked_returns.merge(sp500_returns, left_on='Date', right_on='Date', how='left', suffixes=('', '_SP500'), validate='many_to_one')
combined_returns.tail()
```

```python
final_returns = combined_returns[['TICKER', 'RETURN', 'RETURN_SP500']]
final_returns['EXCESS_RETURN'] = final_returns['RETURN'] - final_returns['RETURN_SP500']
final_returns.tail()
```

```python
## Clean column names
final_returns_cleaned = clean_column_names.clean_column_and_index_names(final_returns)
```

```python
## Save the data to a CSV file for development purposes
final_returns_cleaned.to_csv('../csv_dumps/sp500_returns_time_series.csv')
```

```python
## Get stock info for all S&P 500 stocks (note to exclude S&P 500 Index from the list because the data is returned in a different format)

info_tickers = yf.Tickers(tickers)

info_df = pd.DataFrame()

for ticker in tickers:
    stock_info = info_tickers.tickers[ticker].info
    stock_info_df = pd.DataFrame.from_dict(stock_info, orient='index').T
    info_df = pd.concat([info_df, stock_info_df])

info_df.set_index('symbol', inplace=True)

```

```python
## Clean column names
info_df_cleaned = clean_column_names.clean_column_and_index_names(info_df)
info_df_cleaned.tail()
```
