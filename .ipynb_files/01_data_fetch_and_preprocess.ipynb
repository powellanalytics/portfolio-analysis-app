```python
import yfinance as yf
import pandas as pd
from bs4 import BeautifulSoup
import requests
import datetime
```

```python
request = requests.get(
'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies')
soup = BeautifulSoup(request.content)
main_table = soup.find(id='constituents')
table = main_table.find('tbody').findAll('tr')
table = table[1:]

tickers = []

for row in table:
    ticker = row.findAll('td')[0].text
    tickers.append(ticker)

tickers = [s.replace('\n', '') for s in tickers]

base_url = 'https://en.wikipedia.org'

url_list = []

for item in table:
    url = base_url + str(item.findAll('a')[1]['href'])
    url_list.append(url)

df = pd.DataFrame()
df['tickers'] = tickers
df['url'] = url_list
```

```python
df.tail()
```

```python
start = datetime.datetime(2020, 1, 1)
end = datetime.datetime(2022, 1, 1)
data = yf.download(tickers, start=start, end=end)
data.tail()
```

```python

returns_df = data.stack().reset_index().rename(index=str, columns={"level_1": "Symbol"}).sort_values(['Symbol','Date'])
returns_df.set_index('Date', inplace=True)
returns_df.head(20)
```

```python
returns_df.to_csv('../csv_dumps/sp500_returns.csv')
```

```python
info_tickers = yf.Tickers(tickers)

info_df = pd.DataFrame()

for ticker in tickers:
    stock_info = info_tickers.tickers[ticker].info
    stock_info_df = pd.DataFrame.from_dict(stock_info, orient='index').T
    info_df = pd.concat([info_df, stock_info_df])

info_df.set_index('symbol', inplace=True)

info_df.tail()
```

```python
info_df.to_csv('../csv_dumps/sp500_info.csv')
```
